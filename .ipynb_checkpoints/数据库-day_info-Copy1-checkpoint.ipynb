```python
import cx_Oracle
import pandas as pd

# 设置连接信息
username = 'nppbuf'
password = 'Svwnppbuf321'
hostname = '10.122.6.59'
port = 1521
service_name = 'b2bt'

# 创建连接
dsn = cx_Oracle.makedsn(hostname, port, service_name=service_name)
connection = cx_Oracle.connect(user=username, password=password, dsn=dsn)

# 读取数据
data = {'HOSTPARTID': ['aa', 'aaa', 'aaaa'],
        'HOSTLOCID': ['aa', 'aaa', 'aaaa'],
        'DSHOSTID': ['aa', 'aaa', 'aaaa']}  # 此处添加你的其他列

df = pd.DataFrame(data)

# 定义游标
cursor = connection.cursor()

# 执行插入操作
placeholders = ','.join([':' + str(i+1) for i in range(len(df.columns))])
sql = f"INSERT INTO nppbuf.T_DD_FORECAST_DETAIL_test ({','.join(df.columns)}) VALUES ({placeholders})"
values = [tuple(row) for row in df.to_records(index=False)]

cursor.executemany(sql, values)

# 提交事务
connection.commit()

# 关闭游标和连接
cursor.close()
connection.close()
```
