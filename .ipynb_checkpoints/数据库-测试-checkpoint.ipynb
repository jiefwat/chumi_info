```python
username = 'nppbuf'  
password = 'Svwnppbuf321'  
hostname = '10.122.6.59'  
port = 1521  
service_name = 'b2bt'
dsn = cx_Oracle.makedsn(hostname, port, service_name=service_name)
# 连接 Oracle 数据库  
connection = cx_Oracle.connect(user=username, password=password, dsn=dsn)
# 定义 SQL 查询  
sql_query = """  
     select pm.partnumber,                                 --零件号  
       lm.hostlocid,                                  --仓库代码
       to_char(dh.historybegdate, 'YYYYMMdd'),          --年月
       ds.dsname,                                     --需求流
       dh.historyamount,                              --需求数量
       sa.abc_hits_network,                           --ABC需求频次分类
       sa.abc_cost,                                   --ABC价格分类
       sa.abc_demand,                                 --ABC需求数量分类
       substr(to_char(pm.partnumber), 4, 1)           --零件号第四位
  from nppcore.ipcs_demand_detail dh
  left join nppcore.ipcs_part_master pm
    on dh.partid = pm.partid
  left join nppcore.ipcs_loc_master lm
    on dh.locid = lm.locid
  left join nppcore.ipcs_demand_stream ds
    on dh.DEMANDSTREAMID = ds.DEMANDSTREAMID
  left join nppcore.ipcscust_stock_amount sa
    on dh.partid = sa.partid
   and dh.locid = sa.locid
"""
# 执行 SQL 查询，并按批次抽取数据  
chunksize = 100000  # 每次查询返回的行数，你可以根据需要调整这个值  
query_chunks = pd.read_sql(sql_query, connection, chunksize=chunksize)
# 遍历查询结果的每个批次，并将其合并为一个 DataFrame  
result = pd.DataFrame()  
for chunk in query_chunks: 
    print(chunk.shape)
    result = result.append(chunk)
# 关闭数据库连接  
connection.close()
# 输出合并后的 DataFrame  
result
```
