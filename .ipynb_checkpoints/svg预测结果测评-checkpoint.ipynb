```python
import pandas as pd
# 读取Excel文件的四个sheet
sheet_names = ['001明细数据', '002明细数据', '003明细数据', '004明细数据','异常件18-23']
dfs = []

for sheet_name in sheet_names:
    df = pd.read_csv(f'{sheet_name}.csv',encoding ='gb18030')  # 使用f-string来构建文件名
    df['日期'] = pd.to_datetime(df['日期'], format='%Y%m%d').dt.strftime('%Y-%m-%d')
    df["month"] = pd.to_datetime(df['日期']).dt.month.astype(int)
    df = df[df['日期']>='2023-01-01']
    df['批次']=sheet_name
    dfs.append(df)
pro_kpi_df = pd.concat(dfs, ignore_index=True)
sku_info = pro_kpi_df.groupby(['批次','零件号','month']).size().reset_index(name='sku_mm_values')
month_info = pro_kpi_df.groupby(['批次','month']).size().reset_index(name='mm_values')
result_bety = pd.merge(sku_info,month_info,on = ['批次','month'],how= 'left')
result_bety['Proportion'] = result_bety['sku_mm_values'] / result_bety['mm_values']
result_bety.head()

```

```python
import pandas as pd

# 读取Excel文件的四个sheet
sheet_names = ['rate_001', 'rate_002', 'rate_003', 'rate_004','rate_yichang']
dfs = []

for sheet_name in sheet_names:
    df = pd.read_csv(f'{sheet_name}.csv',encoding ='gb18030')  # 使用f-string来构建文件名
    new_df = df.groupby(['零件号','month','需求频次abc分类'])['real_values','pred_values'].sum().reset_index()
    new_df['批次']=sheet_name
    dfs.append(new_df)

# 将所有DataFrame合并为一个DataFrame
combined_df = pd.concat(dfs, ignore_index=True)
```

```python
poc_yichang = pd.read_csv('svg预测异常件.csv',encoding ='utf-8')
poc_normal = pd.read_csv('svg预测.csv',encoding ='gb18030')
poc_normal_01 = pd.read_csv('01svg预测.csv',encoding ='utf-8')

new_df = poc_yichang.append([poc_normal,poc_normal_01])
new_df['月份'] = pd.to_datetime(new_df['月份'], format='%Y%m').dt.strftime('%Y-%m-%d')
new_df["月份"] = pd.to_datetime(new_df['月份']).dt.month.astype(int)
output =pd.merge(new_df,combined_df,left_on =['零件代码','月份'],right_on =['零件号','month'],how ='left')
output.head()
```

```python
for i in ['预测','real_values']:
    output[i]= [1 if x ==0 else x for x in output[i]]
output['mape'] = (abs(output['real_values'] - output['预测']) / output['real_values']) 
output.head()
```

```python
replacement_dict = {
    '001明细数据': 'rate_001',
    '002明细数据': 'rate_002',
    '003明细数据': 'rate_003',
    '004明细数据': 'rate_004',
    '异常件18-23': 'rate_yichang'
}

result_bety['批次'] = result_bety['批次'].replace(replacement_dict)
result_bety.head()
```

```python
combined_df = pd.merge(output,result_bety[['批次','零件号','month','Proportion','sku_mm_values']],
                       on = ['零件号','month'],how ='left')
combined_df['Proportion'].fillna('0.000001',inplace =True)
combined_df['mape'].fillna('0.000001',inplace =True)
combined_df['Proportion'] = combined_df['Proportion'].astype('float')
combined_df['mape'] = combined_df['mape'].astype('float')

combined_df = combined_df.drop(["批次_y"],axis=1)

combined_df.rename(columns ={'批次_x':'批次'},inplace =True)
```

```python
combined_df.head()
```

```python
# 计算mape,wmape

combined_df['fwmape_values']=combined_df['mape']*combined_df['sku_mm_values']
fwmape_info = combined_df.groupby(['批次','month'])['sku_mm_values','fwmape_values'].sum().reset_index()
fwmape_info.head()
fwmape_info['fwmape'] = 1-  fwmape_info['fwmape_values'] / fwmape_info['sku_mm_values']




mape_info = combined_df.groupby(['批次','month'])['mape'].mean().reset_index()
wmape_info  = combined_df.groupby(['批次','month'])['real_values','预测'].sum().reset_index()
wmape_info['wmape'] = (abs(wmape_info['real_values'] - wmape_info['预测']) / wmape_info['real_values']) 

mape_tur_info = pd.merge(mape_info,wmape_info,on = ['批次','month'],how= 'left')

mape_thr_info = pd.merge(mape_tur_info,fwmape_info,on = ['批次','month'],how= 'left')

mape_thr_info.head()
```

```python
mape_thr_info.to_csv('111.csv',encoding ='gb18030')
```

```python
#计算需求频次abc分类mape，wmape
grouped_mape_info = combined_df.groupby(['month','需求频次abc分类'])['mape'].mean().reset_index()
grouped_wmape_info  = combined_df.groupby(['需求频次abc分类','month'])['real_values','预测'].sum().reset_index()
grouped_wmape_info['wmape'] = (abs(grouped_wmape_info['real_values'] - grouped_wmape_info['预测']) / grouped_wmape_info['real_values'])
grouped_mape_tur_info = pd.merge(grouped_mape_info,grouped_wmape_info,on = ['需求频次abc分类','month'],how= 'left')

grouped_fwmape_info = combined_df.groupby(['需求频次abc分类','month'])['sku_mm_values','fwmape_values'].sum().reset_index()
grouped_fwmape_info['fwmape'] = 1-  grouped_fwmape_info['fwmape_values'] / grouped_fwmape_info['sku_mm_values']

group_mape_thr_info = pd.merge(grouped_mape_tur_info,grouped_fwmape_info,on = ['需求频次abc分类','month'],how= 'left')

group_mape_thr_info.head()
```

```python
group_mape_thr_info.to_csv('112.csv',encoding = 'gb18030')
```

```python
# 计算mape，wmape 占比情况

# 定义区间
import matplotlib.pyplot as plt
import seaborn as sns

bins = [0, 0.10, 0.30, 0.50, 1.00, 100000]

# 使用cut函数将数据分成不同区间
combined_df['ScoreRange'] = pd.cut(combined_df['mape'], bins=bins, labels=['0-10%', '10%-30%', '30%-50%', '50%-100%', '>100%'],right =False)

combined_df['ScoreRange'].replace('NaN','0',inplace =True)
# 计算每个区间中的数据数量

grouped = combined_df.groupby(['month','ScoreRange']).size().reset_index(name='sku_mm_values')
# 计算每个月份内的总样本数
monthly_counts = combined_df.groupby(['month']).size().reset_index(name='mm_values')

# 合并统计结果
result = pd.merge(grouped, monthly_counts, on=['month'],how = 'left')

# 计算占比
result['Proportion'] = result['sku_mm_values'] / result['mm_values']
result.tail(10)
```

```python
result.to_csv('113.csv',encoding ='gb18030')
```
